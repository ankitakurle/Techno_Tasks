### Task 3 : Visualization using Histogram
#### Create a histogram or bar chart to visualize the distribution of data in a dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/0863a/Downloads/archive/Iris.csv", encoding= 'unicode_escape')
```

```python
data
```

```python
## bar plot for sepal length
ax = sns.countplot(x = 'SepalLengthCm',data = data)
sns.set(rc={'figure.figsize':(25,10)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
## barplot for petal length
ax = sns.countplot(x = 'PetalLengthCm',data = data)
sns.set(rc={'figure.figsize':(25,10)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
## barplot for petak width
ax = sns.countplot(x = 'PetalWidthCm',data = data)
sns.set(rc={'figure.figsize':(25,10)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
## barplot for sepalwidth
ax = sns.countplot(x = 'SepalWidthCm',data = data)
sns.set(rc={'figure.figsize':(25,10)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# barplot for sepal length and width
sap_ = data.groupby(['SepalLengthCm'], as_index=False)['SepalWidthCm'].sum().sort_values(by='SepalWidthCm', ascending=False).head(10)

sns.barplot(x = 'SepalLengthCm',y= 'SepalWidthCm' ,data = sap_)
```

```python
## barplot for petal length and width
pet_ = data.groupby(['PetalLengthCm'], as_index=False)['PetalWidthCm'].sum().sort_values(by='PetalWidthCm', ascending=False).head(10)

sns.barplot(x = 'PetalLengthCm',y= 'PetalWidthCm' ,data = pet_)
```
